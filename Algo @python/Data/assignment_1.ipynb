implement the Gradient Descent algorithm on a toy 2-D dataset

```python
 #Sigmoid neuron
import math

def func(w,b,x):
    return 1/(1+math.exp(-(w*x+b)))
```

```python
# returns squared error loss
def error(w,b):
    err=0
    for x,y in zip(X,Y):
        fx=func(w,b,x)
        err=err + 0.5*((fx-y)**2)
    return err
```

```python
def delta_w(w,b,x,y):
    fx= func(w,b,x)
    return (fx-y)*fx*(1-fx)*x
```

```python
def delta_b(w,b,x,y):
    fx= func(w,b,x)
    return (fx-y)*fx*(1-fx)
```

```python
def vanilla_GD():
    w=1
    b=1
    eta= 0.01
    iterations=6000
    
    for i in range(iterations):
        dw, db =0,0

        #Calculate delta w ,b for all points
        for x,y in zip(X,Y):
            dw = dw + delta_w(w,b,x,y)
            db = db + delta_b(w,b,x,y)
    
        w=w - eta*dw
        b=b - eta*db
        
        print("epoch = {0} : loss = {1}".format(i,error(w,b)))
        
    return (w,b)

```

```python
import pandas as pd
data = pd.read_csv("A2_Q4_data.csv")
X = data["X"]
Y = data["Y"]
w,b = vanilla_GD()
print("final Error= ",error(w,b))
```
